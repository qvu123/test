```python
# Lấy dữ liệu theo năm 2017, 2018, tháng của một thành phố theo bệnh
def get_disease_by_year_month(city_name='', disease_name ='' , year = '', month =''):     
    if disease_name == 'DH':
        if year == 2017:
            disease_city_value = dh_data_2017.loc[ (dh_data_2017['province'] == city_name)]       
        if year == 2018:
            disease_city_value = dh_data_2018.loc[ (dh_data_2018['province'] == city_name)]     
    if disease_name == 'DF':
        if year == 2017:
            disease_city_value = df_data_2017.loc[ (df_data_2017['province'] == city_name)]       
        if year == 2018:
            disease_city_value = df_data_2018.loc[ (df_data_2018['province'] == city_name)] 
        if year == 2019: # 2019 hiện chỉ có DF
            disease_city_value = df_data_2019.loc[ (df_data_2019['province'] == city_name)]  
    if disease_name == 'INF':
        if year == 2017:
            disease_city_value = inf_data_2017.loc[ (inf_data_2017['province'] == city_name)]       
        if year == 2018:
            disease_city_value = inf_data_2018.loc[ (inf_data_2018['province'] == city_name)]  
    # DC = death cases
    if disease_name == 'DH_DC':
        if year == 2017:
            disease_city_value = dh_data_2017_death_case.loc[ (dh_data_2017_death_case['province'] == city_name)]       
        if year == 2018:
            disease_city_value = dh_data_2018_death_case.loc[ (dh_data_2018_death_case['province'] == city_name)]       
    if disease_name == 'DF_DC':
        if year == 2017:
            disease_city_value = df_data_2017_death_case.loc[ (df_data_2017_death_case['province'] == city_name)]       
        if year == 2018:
            disease_city_value = df_data_2018_death_case.loc[ (df_data_2018_death_case['province'] == city_name)]  
        if year == 2019:
            disease_city_value = df_data_2019_death_case.loc[ (df_data_2019_death_case['province'] == city_name)]
    if disease_name == 'INF_DC':
        if year == 2017:
            disease_city_value = inf_data_2017_death_case.loc[ (inf_data_2017_death_case['province'] == city_name)]       
        if year == 2018:
            disease_city_value = inf_data_2018_death_case.loc[ (inf_data_2018_death_case['province'] == city_name)]        

    return disease_city_value[month].values[0]
```

```python
def merge_climate_factor_in_month(one_month_city, column_name =''):
    mean_val = pd.to_numeric(one_month_city[column_name], errors='coerce').mean()
    return mean_val
```

```python
def calculate_incident_rate(one_month_city):
    incident_rate = {}
    province_name = one_month_city.iloc[0]['province']
    year = one_month_city.iloc[0]['year']
    population_of_city  = get_pop_city_year(city_name= province_name,year = year)   
    if population_of_city == '..': # nếu không có dữ liệu dân số thì để trống
        incident_rate['Influenza_rates'] = np.nan
        incident_rate['Dengue_fever_rates'] = np.nan
        incident_rate['Diarrhoea_rates'] = np.nan
    else: # nếu có dữ liệu dân số thì tính
        # dữ liệu dân số rút gọn cần x 1000 mới ra kết quả.
        # incident = ( case/ dân số *1000) x100.000
        incident_rate['Influenza_rates'] = (one_month_city.iloc[0]['Influenza cases'] / population_of_city)* 100
        incident_rate['Dengue_fever_rates'] = (one_month_city.iloc[0]['Dengue fever-cases'] / population_of_city)* 100
        incident_rate['Diarrhoea_rates'] = (one_month_city.iloc[0]['Diarrhoea-cases'] / population_of_city)* 100
    
    return incident_rate
```

```python
def merge_month_data(one_month):
    incident_rate = calculate_incident_rate(one_month)
    area_crr, population_density_crr, population_average_crr = get_population_area_density(one_month)
    df = pd.DataFrame({
                        'year': one_month.iloc[0]['year'],
                        'month': one_month.iloc[0]['month'],
                        'province': one_month.iloc[0]['province'],
                        'Influenza_cases': one_month.iloc[0]['Influenza cases'],
                        'Dengue_fever_cases': one_month.iloc[0]['Dengue fever-cases'],
                        'Diarrhoea_cases': one_month.iloc[0]['Diarrhoea-cases'],
                        'Total_Evaporation': merge_climate_factor_in_month(one_month,'Total monthly evaporation (mm)'),
                        'Total_Rainfall': merge_climate_factor_in_month(one_month,'Total monthly rainfall (mm)'),
                        'Max_Daily_Rainfall': merge_climate_factor_in_month(one_month,'The highest daily total rainfall for the month (mm)'),
                        'n_raining_days': merge_climate_factor_in_month(one_month,'Total number of monthly rainning_days (day)'),
                        'Average_temperature': merge_climate_factor_in_month(one_month,'Monthly average temperature °C'),
                        'Max_Average_Temperature': merge_climate_factor_in_month(one_month,'Max monthly average temp °C'),
                        'Min_Average_Temperature': merge_climate_factor_in_month(one_month,'Min montlhy average temp °C'),
                        'Max_Absolute_Temperature': merge_climate_factor_in_month(one_month,'Monthly absolute maximum temperature (°C)'),
                        'Min_Absolute_Temperature': merge_climate_factor_in_month(one_month,'Monthly absolute  minimum temperature (°C)'),
                        'Average_Humidity': merge_climate_factor_in_month(one_month,'Monthly average relative humidity (%)'),
                        'Min_Humidity': merge_climate_factor_in_month(one_month,'Monthly minimal relative humudity (%)'),
                        'n_hours_sunshine': merge_climate_factor_in_month(one_month,'Total hours of sunshine (hours)'),
                        # Đoạn này thêm vào dữ liệu thống kê dân số theo tỉnh
                        'population_male': population_male.loc[one_month.iloc[0]['province'], str(one_month.iloc[0]['year'])],
                        'population_female': population_female.loc[one_month.iloc[0]['province'], str(one_month.iloc[0]['year'])],
                        'population_city': population_city.loc[one_month.iloc[0]['province'], str(one_month.iloc[0]['year'])],
                        'population_countryside': population_countryside.loc[one_month.iloc[0]['province'], str(one_month.iloc[0]['year'])],
                        
                        # Đoạn này thêm vào dữ liệu diện tích và mật độ 2011 -2020                        
                        'area': area_crr,
                        'population_density': population_density_crr,
                        'population_average': population_average_crr,

                        # Đoạn này thêm vào dữ liệu tỉ lệ sinh đẻ                    
                        'birth_rate': get_birth_rate(one_month),

                        # Đoạn này thêm vào dữ liệu nguồn nước                   
                        'urban_water_usage_rate': get_urban_water_usage(one_month), 
                        'clean_water_rate_all': get_clean_water_rate_all(one_month), 

                        # Đoạn này thêm vào dữ liệu nghèo               
                        'poverty_rate': get_poverty_rate(one_month),   

                        # Đoạn này thêm vào dữ liệu toilet               
                        'toilet_rate': get_toilet_rate(one_month),                       

                        'Influenza_rates': incident_rate['Influenza_rates'],
                        'Dengue_fever_rates': incident_rate['Dengue_fever_rates'],
                        'Diarrhoea_rates': incident_rate['Diarrhoea_rates']
                    },index=[0])
    
    return df
```
